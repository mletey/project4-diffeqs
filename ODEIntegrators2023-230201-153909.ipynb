# ODE

Equation: Solve harmonic oscillator

```julia
# ẍ = -x
```

```julia
function harmonic(y)
    x, v = y
    ẋ = v
    v̇ = -x
    ẏ = [ẋ, v̇]
    return ẏ
end
```

```julia
function euler(f, y₀, h)
    k₀ = f(y₀)
    y₁ = y₀ + h * k₀
    return y₁
end
```

```julia
y₀ = [1.0, 0.0]
```

```julia
euler(harmonic, y₀, 0.1)
```

```julia
function evolve(f, y, h, nsteps)
    trajectory = [y]
    for n in 1:nsteps
        y = euler(f, y, h)
        push!(trajectory, y)
    end
    return trajectory
end
```

```julia
ys = evolve(harmonic, y₀, 0.1, 100);
```

```julia
ys2 = evolve(harmonic, y₀, 0.05, 200);
```

```julia
using WGLMakie
```

```julia
fig = Figure()
ax = Axis(fig[1, 1])
plot!(map(y -> y[1], ys), map(y -> y[2], ys); color=:red)
plot!(map(y -> y[1], ys2), map(y -> y[2], ys2); color=:green)
fig
```

```julia
function midpoint(f, y₀, h)
    k₀ = f(y₀)
    y₁ = y₀ + h/2 * k₀
    k₁ = f(y₁)
    y₂ = y₀ + h * k₁
    return y₂
end
```

```julia
function evolve_midpoint(f, y, h, nsteps)
    trajectory = [y]
    for n in 1:nsteps
        y = midpoint(f, y, h)
        push!(trajectory, y)
    end
    return trajectory
end
```

```julia
ys_m = evolve_midpoint(harmonic, y₀, 0.1, 100);
```

```julia
fig = Figure()
ax = Axis(fig[1, 1])
plot!(map(y -> y[1], ys), map(y -> y[2], ys); color=:red)
plot!(map(y -> y[1], ys_m), map(y -> y[2], ys_m); color=:green)
fig
```

```julia
ys_m1 = evolve_midpoint(harmonic, y₀, 0.5, 20);
ys_m2 = evolve_midpoint(harmonic, y₀, 0.25, 40);
```

```julia
fig = Figure()
ax = Axis(fig[1, 1])
lines!(map(y -> y[1], ys_m1), map(y -> y[2], ys_m1); color=:red)
lines!(map(y -> y[1], ys_m2), map(y -> y[2], ys_m2); color=:green)
fig
```
